```python
two_category_mapping = {
    "Action": "Speculative",
    "Adventure": "Speculative",
    "Avant Garde": "Speculative",
    "Award Winning": "Realistic",
    "Boys Love": "Realistic",
    "Comedy": "Realistic",
    "Drama": "Realistic",
    "Ecchi": "Speculative",
    "Erotica": "Speculative",
    "Fantasy": "Speculative",
    "Girls Love": "Realistic",
    "Gourmet": "Realistic",
    "Horror": "Speculative",
    "Mystery": "Speculative",
    "Romance": "Realistic",
    "Sci-Fi": "Speculative",
    "Slice of Life": "Realistic",
    "Sports": "Realistic",
    "Supernatural": "Speculative",
    "Suspense": "Speculative"
}

animes["simple_genres"] = animes["Genres"].map(two_category_mapping)
```

```python
from transformers import pipeline

genre_categories = ["Realistic", "Speculative"]
pipe = pipeline("zero-shot-classification",
                model="valhalla/distilbart-mnli-12-1",
                device = 0)

```

```python
sequence = animes.loc[animes["simple_genres"] == "Realistic", "description"].reset_index(drop=True)[0]
```

```python
pipe(sequence, genre_categories)
```

```python
import numpy as np

max_index = np.argmax(pipe(sequence, genre_categories)["scores"])
max_label = pipe(sequence, genre_categories)["labels"][max_index]
max_label
```

```python
def generate_predictions(sequence, categories):
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm

actual_cats = []
predicted_cats = []

for i in tqdm(range(0, 300)):
    sequence = animes.loc[animes["simple_genres"] == "Realistic", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, genre_categories)]
    actual_cats += ["Realistic"]
```

```python
for i in tqdm(range(0, 300)):
    sequence = animes.loc[animes["simple_genres"] == "Speculative", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, genre_categories)]
    actual_cats += ["Speculative"]
```

```python
predictions_df = pd.DataFrame({"actual_categories": actual_cats, "predicted_categories": predicted_cats})
```

```python
predictions_df["correct_prediction"] = (
    np.where(predictions_df["actual_categories"] == predictions_df["predicted_categories"], 1, 0)
)
```

```python
ids = []
prediction_cats = []

missing_cats = animes.loc[animes["simple_genres"].isna(), ["myanimelist_id", "description"]].reset_index(drop=True)
```

```python
def generate_predictions2(sequences, categories):
    outputs = pipe(sequences, categories)  # returns list if sequences is a list

    results = []
    for pred in outputs:
        max_index = np.argmax(pred["scores"])
        max_label = pred["labels"][max_index]
        results.append(max_label)

    return results
```

```python
batch_size = 20

for start in tqdm(range(0, len(missing_cats), batch_size)):
    batch = missing_cats.iloc[start:start + batch_size]

    descriptions = batch["description"].tolist()
    response = generate_predictions2(descriptions, genre_categories)

    prediction_cats.extend(response)
    ids.extend(batch["myanimelist_id"].tolist())
```

```python
missing_predicted_df = pd.DataFrame({"ids": ids, "predicted_categories": predicted_cats})
```

```python
missing_predicted_df
```

```python
animes = pd.merge(animes, missing_predicted_df, on="ids", how="left")
animes["simple_genres"] = np.where(animes["simple_genres"].isna(), animes["predicted_categories"], animes["simple_genres"])
animes = animes.drop(columns = ["predictied_categories"])
```
