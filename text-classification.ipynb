```python
two_category_mapping = {
    "Action": "Speculative",
    "Adventure": "Speculative",
    "Avant Garde": "Speculative",
    "Award Winning": "Realistic",
    "Boys Love": "Realistic",
    "Comedy": "Realistic",
    "Drama": "Realistic",
    "Ecchi": "Speculative",
    "Erotica": "Speculative",
    "Fantasy": "Speculative",
    "Girls Love": "Realistic",
    "Gourmet": "Realistic",
    "Horror": "Speculative",
    "Mystery": "Speculative",
    "Romance": "Realistic",
    "Sci-Fi": "Speculative",
    "Slice of Life": "Realistic",
    "Sports": "Realistic",
    "Supernatural": "Speculative",
    "Suspense": "Speculative"
}

animes["simple_genres"] = animes["Genres"].map(two_category_mapping)
```

```python
from transformers import pipeline
genre_categories = ["Realistic", "Speculative"]
pipe = pipeline("zero-shot-classification",
                model="typeform/distilbert-base-uncased-mnli",
                device=0)

```

```python
sequence = animes.loc[animes["simple_genres"] == "Realistic", "description"].reset_index(drop=True)[0]
```

```python
pipe(sequence, genre_categories)
```

```python
import numpy as np

max_index = np.argmax(pipe(sequence, genre_categories)["scores"])
max_label = pipe(sequence, genre_categories)["labels"][max_index]
max_label
```

```python
def generate_predictions(sequence, categories):
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm

actual_cats = []
predicted_cats = []

for i in tqdm(range(0, 300)):
    sequence = animes.loc[animes["simple_genres"] == "Realistic", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, genre_categories)]
    actual_cats += ["Realistic"]
```

```python
for i in tqdm(range(0, 300)):
    sequence = animes.loc[animes["simple_genres"] == "Speculative", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, genre_categories)]
    actual_cats += ["Speculative"]
```

```python
predictions_df = pd.DataFrame({"actual_categories": actual_cats, "predicted_categories": predicted_cats})
```

```python
predictions_df
```

```python
predictions_df["correct_prediction"] = (
    np.where(predictions_df["actual_categories"] == predictions_df["predicted_categories"], 1, 0)
)
```

```python
predictions_df["correct_prediction"].sum() / len(predictions_df)
```

```python
ids = []
prediction_cats = []

missing_cats = animes.loc[animes["simple_genres"].isna(), ["myanimelist_id", "description"]].reset_index(drop=True)
```

```python
n = len(missing_cats) // 8
print(n)

for i in tqdm(range(0, n)):
    sequence = missing_cats["description"][i]  # ✓ Get description for row i
    prediction_cats.append(generate_predictions(sequence, genre_categories))  # No need for extra list wrap
    ids.append(missing_cats["myanimelist_id"][i])  # No need for extra list wrap
```

```python
# Create DataFrame
missing_predicted_df = pd.DataFrame({
    "ids": ids,
    "predicted_categories": prediction_cats
})
```

```python
missing_predicted_df
```

```python
# Flatten the lists in ids column
missing_predicted_df['ids'] = missing_predicted_df['ids'].apply(
    lambda x: x[0] if isinstance(x, list) else x
)

# Convert to int
missing_predicted_df['ids'] = missing_predicted_df['ids'].astype(int)

# Flatten predicted_categories if they're also nested lists
if isinstance(missing_predicted_df['predicted_categories'].iloc[0], list):
    missing_predicted_df['predicted_categories'] = missing_predicted_df['predicted_categories'].apply(
        lambda x: x[0] if isinstance(x, list) and len(x) > 0 else x
    )

# Now merge
animes = pd.merge(animes, missing_predicted_df, left_on="myanimelist_id", right_on="ids", how="left")

# Fill missing simple_genres with predicted_categories
animes["simple_genres"] = np.where(animes["simple_genres"].isna(), animes["predicted_categories"], animes["simple_genres"])

# Drop the extra columns
animes = animes.drop(columns=["ids", "predicted_categories"])

print("Merge successful!")
print(f"Total rows: {len(animes)}")
print(f"Rows with simple_genres filled: {animes['simple_genres'].notna().sum()}")
```
