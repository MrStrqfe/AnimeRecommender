```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

animes = pd.read_csv("animes_cleaned.csv")
```

```python
animes
```

```python
animes["tagged_description"].to_csv("tagged_description.txt",
                                   index=False,
                                   header=False)
```

```python
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
db_animes = Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings()
)
```

```python
query = "A fantasy anime with magic and adventure"
docs = db_animes.similarity_search(query, k = 10)
docs
```

```python
anime_id_str = docs[0].page_content.split()[0].strip('"')
anime_id = int(anime_id_str)
anime_row = animes[animes["myanimelist_id"] == anime_id]
anime_row
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_animes.similarity_search(query, k = 50)

    animes_list = []

    for i in range(0, len(recs)):
        animes_list += [int(recs[i].page_content.strip('"').split()[0])]

    return animes[animes["myanimelist_id"].isin(animes_list)].head(top_k)
```

```python
retrieve_semantic_recommendations("A fantasy anime with magic and adventure")
```
